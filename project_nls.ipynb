```python
ni = [800, 900, 1000]

ne = [200, 400, 600, 800, 1000]
```

```python
neuron_dict = {}

for i, n_i in enumerate(ni):
    for j, n in enumerate(ne):
        neuron_dict[i,j] = n_i, n

list_spikes= list(neuron_dict.values())

list_spikes
```

```python

import matplotlib.pyplot as plt

# fixo um ni e vario os nes


#regularization = [0.0001, 0.005, 0.1, 0.5]

#coefs = ['w0','w1','w2','w3','w4','w5','w6','w7','w8','w9']

fig, axs = plt.subplots(1,5)
fig.set_size_inches(20, 20)
ind = 0

for ax in axs.flat:

    times, spikes = main(ni[ind],ne[ind])

    ax.set_title(f'Ni={ni[ind]}, Ne={ne[ind]}')

    ax.plot(times,spikes, ',k')
        
    ind += 1

        

```

```python
for i in range(0, len(list_spikes)):
    ni = list_spikes[i][0]
    ne = list_spikes[i][1]
    
    fig, axs = plt.subplots(1,5)
    fig.set_size_inches(20, 20)
    ind = 0

    for ax in axs.flat:

        times, spikes = main(ni,ne)

        ax.set_title(f'Ni={ni[ind]}, Ne={ne[ind]}')

        ax.plot(times,spikes, ',k')
            
        ind += 1
```
